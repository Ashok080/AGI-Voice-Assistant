```python
# ‚¨áÔ∏è AGI Voice Agent - Dataset Downloader (Notebook 4)

### Purpose:
Automatically download all 73 datasets from:
- Google Drive (via gdown)
- OR S3 Bucket
- OR Supabase Storage

This ensures your AGI system can bootstrap datasets on any machine or cloud VM.





üìå 1. Install Required Libraries

!pip install gdown boto3 supabase




üìå 2. Define Dataset Sources

dataset_links = {
    "AI_Reasoning": "https://drive.google.com/uc?id=FILE_ID1",
    "Hellaswag_Test": "https://drive.google.com/uc?id=FILE_ID2",
    "Hellaswag_Val": "https://drive.google.com/uc?id=FILE_ID3",
    # Continue for all 73 datasets
}




üìå 3. Download from Google Drive (gdown)

import gdown
from pathlib import Path

Path("data").mkdir(exist_ok=True)

def download_from_drive(name, url):
    output = f"data/{name}.downloaded"
    gdown.download(url, output, quiet=False)
    print(f"‚úÖ Downloaded {name} to {output}")

for name, link in dataset_links.items():
    download_from_drive(name, link)




üìå 4. (Optional) Download from AWS S3

If your datasets are in S3:

import boto3

s3 = boto3.client('s3')

def download_from_s3(bucket, s3_key, local_path):
    s3.download_file(bucket, s3_key, local_path)
    print(f"‚úÖ Downloaded from S3: {local_path}")

# Example
# download_from_s3('my-bucket', 'datasets/dataset1.csv', 'data/dataset1.csv')




üìå 5. (Optional) Download from Supabase Storage

from supabase import create_client

url = 'YOUR_SUPABASE_URL'
key = 'YOUR_SUPABASE_KEY'
supabase = create_client(url, key)

def download_from_supabase(bucket, path, save_as):
    response = supabase.storage.from_(bucket).download(path)
    with open(f"data/{save_as}", 'wb') as f:
        f.write(response)
    print(f"‚úÖ Downloaded {save_as} from Supabase.")

# Example
# download_from_supabase('datasets', 'dataset1.csv', 'dataset1.csv')




üìå 6. Verify All Files Downloaded

import os

for name in dataset_links:
    file_path = f"data/{name}.downloaded"
    if os.path.exists(file_path):
        print(f"‚úÖ {name} is downloaded.")
    else:
        print(f"‚ùå {name} is missing.")
```
